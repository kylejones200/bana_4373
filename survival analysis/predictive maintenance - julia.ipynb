```julia
using PrettyPrinting
using DataFrames, CSV
using MLJ
```

```julia
target = "data/machines.csv"

df = CSV.read(target);
```

```julia
describe(df)
```

```julia
cols_to_drop = [:Column1, :source, :id, :Category, :max]
df = select!(df, Not(cols_to_drop))
```

```julia
df = DataFrame(df)
```

```julia
y, X = unpack(df, ==(:RUL), colname->true);
```

```julia
X = DataFrame(X)
```

```julia
describe(X)
```

```julia
df = select!(df, Not([:Category, :survival]))
```

```julia
@load PCA

pca = X  |> PCA(maxoutdim=5)

```

```julia
fit!(pca)
```

```julia
fit!(pca)
W = pca();
```

```julia
using MultivariateStats

transform(W, X)
```

```julia
p = fit(PCA,X)
```

```julia
W.category = df.Category
```

```julia
W
```

```julia
xgreg = XGBoostCount()

xgmach = machine(xgreg, W, y)

train, test = partition(eachindex(y), 0.70, shuffle=true); # 70:30 split
fit!(xgmach, rows=train)
ŷ = predict(cmach, rows=test)
ŷ[1:3] |> pprint
```

```julia
@load RidgeRegressor pkg="MultivariateStats"
@load KNNRegressor
```

```julia
Xs = source(W)
ys = source(y, kind=:target)
```

```julia
hot = machine(OneHotEncoder(), Xs)

W = transform(hot, Xs)
z = log(ys);
```

```julia
knn   = machine(KNNRegressor(K=5), W, z)
ridge = machine(RidgeRegressor(lambda=2.5), W, z)

ẑ₁ = predict(ridge, W)
ẑ₂ = predict(knn, W)
```
