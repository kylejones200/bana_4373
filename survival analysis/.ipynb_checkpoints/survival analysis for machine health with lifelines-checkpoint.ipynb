```python
import pandas as pd

%matplotlib inline
```

```python
file = "data/predictive maintenance trainALL.csv"

df = pd.read_csv(file)
```

```python
df.describe()
```

```python
df.head()
```

```python
max_rul = df.groupby(['source','id'])['max'].max()
```

```python
len(max_rul)
```

```python
max_rul = pd.DataFrame(max_rul)
```

```python
max_rul["observed"] = 1
```

```python
max_rul.head()
```

```python
# Python code to create the above Kaplan Meier curve
from lifelines import KaplanMeierFitter

## Example Data 
durations = max_rul["max"]
event_observed = max_rul["observed"]

## create a kmf object
kmf = KaplanMeierFitter() 

## Fit the data into the model
kmf.fit(durations, event_observed,label='Kaplan Meier Estimate')

## Create an estimate
kmf.plot(ci_show=False) ## ci_show is meant for Confidence interval, since our data set is too tiny, thus i am not showing it.
```

```python
kmf.plot(at_risk_counts=True)
```

```python
import lifelines

lifelines.utils.qth_survival_time(.5, kmf)
```

```python
from lifelines.datasets import load_rossi
from lifelines import CoxPHFitter

cph = CoxPHFitter()
cph.fit(max_rul, duration_col='max', event_col='observed')

cph.plot_partial_effects_on_outcome('max', [0, 100, 200, 300, 400, 500], cmap='coolwarm')
```

```python
cph.print_summary()
```
