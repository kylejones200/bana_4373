```python
import pandas as pd
```

```python
df = pd.read_csv("power.csv")
```

```python
df.head()
```

```python
df["D"] = pd.to_datetime(df["D"])

df.set_index("D", inplace=True)
```

```python
df.head()
```

```python
df.drop(["date"], axis=1, inplace=True)
```

```python
df["diff"] = df["cum_power"].diff()
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
df.to_csv("power.csv")
```

```python
df["diff"].plot()
```
