```python
import pandas as pd
import seaborn as sns
%matplotlib inline

df = pd.read_excel("data/froggy1.xlsx")
```

```python
import missingno as msno

msno.matrix(df)
```

```python
df_missing = df.dropna()
```

```python
msno.matrix(df_missing)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
j = imp_mean.fit_transform(df["angle_01"].values.reshape(-1,1))
j = pd.DataFrame(j)
```

```python
df["angle_01"].hist()
```

```python
j.hist()
```

```python
sns.displot(j, x=0)
```
