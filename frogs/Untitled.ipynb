```python
import plotly.express as px


fig = px.violin(df, y="distance",  color="frog_type", box=True, points="all",
          hover_data=df.columns, title="How far can a frog jump?")

fig.update_layout(shapes=[
dict(
  type= 'line',
  yref= 'y', y0= df["distance"].mean(), y1= df["distance"].mean(),   # adding a horizontal line at Y = 1
  xref= 'paper', x0= 0, x1= 1
     ) 
])

fig.show()
fig.write_html("frogDist_by_type.html")
```

```python
import plotly.figure_factory as ff
import numpy as np


group_labels = ['Pro', 'Rental', 'Unknown', 'Individual']

colors = ['blue', 'red', 'green', 'purple']

# Create distplot with curve_type set to 'normal'
fig = ff.create_distplot(df, y="distance", group_labels, bin_size=5,
                         curve_type='normal', # override default 'kde'
                         colors=colors)

# Add title
fig.update_layout(title_text='Distplot with Normal Distribution')
fig.show()
```

```python
fig = ff.create_distplot([df[c] for c in df.columns], df.columns, bin_size=.25)
fig.show()
```

```python
import plotly.express as px


fig = px.histogram(df, x="distance", color="frog_type",
                   marginal="box", # or violin, rug
                   hover_data=df.columns)
fig.show()
```
