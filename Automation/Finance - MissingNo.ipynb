# Identifying missing values 

Data completeness is key to data integrity but it can be hard to identify if a dataset is complete. The module "MissingNo" makes a simple visualization to see where data is missing, signified in white.



```python
import pandas as pd
import missingno as msno
%matplotlib inline

df = pd.read_csv('shell financial data.csv')
msno.matrix(df)
```

# Handling missing data

There are two general options for handling missing values: dropping observations with missing data or imputing missing data.

## Dropping missing values
If you have a large dataset, then dropping incomplete observations is often the best choice. 

## Imputing missing values
There are several methods for imputing missing values including missing in missing values with a constant value, the mean or median of the avialable values, or the result of an ML method such as linear discriminate analysis. 

Pandas and SciKit-Learn offer methods for imputing missing values.
